```python
import pandas as pd

PATH = './Database.xlsx'
df   = pd.read_excel(PATH)
df   = df.drop(columns=['Receita Média por Compra', 'Plataforma Utilizada', 'Gênero', 'Data', 'Cidade'])
df   = df.dropna() # drop all those rows which have any 'nan' value in it.
df   = df.drop_duplicates() # Remove duplicates
df   = df.rename(columns={'Receita Total': 'Receita_Total'})
df   = df.astype({'Receita_Total': 'int64', 'Número de Compradores': 'int64', 'Novos Usuários': 'int64', 
                  'Visualizações na Página': 'int64', 'Usuários Ativos': 'int64', 'Usuários Total': 'int64'})
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.bar(df['Origem'], df['Receita_Total'])

ax.set_title('Gráfico de Barras')
ax.set_xlabel('Origem')
ax.set_ylabel('Receita')
```

```python
fig = plt.figure(figsize=(9,4))

x = df['Origem']
y = df['Receita_Total']

plt.bar(x, y)
plt.title('Gráfico de Barras')
plt.xlabel('Origem')
plt.ylabel('Receita')


colors = ['purple', 'orange', 'yellow']
plt.bar(x, y, color=colors)
plt.show()
```
